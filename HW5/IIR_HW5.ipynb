# Hypothesis testing #1
#### 設:
- H0: 婚姻 != 平均滿意度
- H1: 婚姻 = 平均滿意度
- 信心水準: 95%
- 統計檢定方法: t-test

#### 結果:
- p-value: 0.822
- 推論: 拒絕H1，婚姻與平均滿意度無相關

#### 程式碼:


```python
sample['marriage'].unique()
```

```python
from scipy.stats import ttest_ind

group1 = sample[sample['marriage']==1]['平均滿意度']
group2 = sample[sample['marriage']==2]['平均滿意度']

statval,pval = ttest_ind(group1,group2)
print("Test statistic:",statval)
print("P-value:",pval)
if pval <0.05:
  print("拒絕H0,有相關")
else:
  print("拒絕H1,無相關")

```

# Hypothesis testing #2
#### 設:
- H0: 地區 != 平均滿意度
- H1: 地區 = 平均滿意度
- 信心水準: 95%
- 統計檢定方法: ANOVA

#### 結果:
- p-value: 0.000002 
- 推論: 拒絕H0，地區與平均滿意度有關

#### 程式碼:

```python
sample['Area'].unique()
```

```python
from scipy.stats import f_oneway

group1 = sample[sample['Area']==1]['平均滿意度']
group2 = sample[sample['Area']==2]['平均滿意度']
group3 = sample[sample['Area']==3]['平均滿意度']
group4 = sample[sample['Area']==4]['平均滿意度']

statval,pval = f_oneway(group1, group2, group3, group4)
print("Test statistic:",statval)
print("P-value:",pval)
if pval <0.05:
  print("拒絕H0,有相關")
else:
  print("拒絕H1,無相關")
```

# Hypothesis testing #3
#### 設:
- H0: 年齡 != Q3(我對軟體問題的詢問，能有快速的回應)
- H1: 年齡 = Q3(我對軟體問題的詢問，能有快速的回應)
- 信心水準: 95%
- 統計檢定方法: Pearson's Chi-square test 

#### 結果:
- p-value: 0.0014
- 推論: 拒絕H0，年齡與Q3(我對軟體問題的詢問，能有快速的回應)有關

#### 程式碼:

```python
sample['age'].unique()
```

```python
sample['Q3'].unique()
```

```python
from scipy.stats import chi2_contingency

CrossTable = pd.crosstab(sample['age'], sample['Q3'])
statval,pval,dof,ex = chi2_contingency(CrossTable)
print("Test statistic:",statval)
print("P-value:",pval)
if pval <0.05:
  print("拒絕H0,有相關")
else:
  print("拒絕H1,無相關")
```
